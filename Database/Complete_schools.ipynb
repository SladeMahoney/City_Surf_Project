```python
# Drop columns we do not want in schema
schools_df = schools_df.drop(columns = ["Contract ID", "Agency Name", "Workplan Activity ID", 
                                        "Activity Type", "Activity Description", "Service Site Zip", 
                                        "Lead Staff", "Service Site Phone","Service Site Location", "Service Site Id"])
schools_df
```

```python
# Are all Intro to surfing field trips one day?
intro_surfing_field_trips = schools_df.loc[(schools_df["Activity Name"]=="Intro to Surfing Field Trips")]
intro_surfing_field_trips
```

```python
# Fixing Datatypes
# Convert Date of Birth and agency enrollment date to Datetime
schools_df['Start Date'] = pd.to_datetime(schools_df['Start Date'])
schools_df['End Date'] = pd.to_datetime(schools_df['End Date'])
```

```python
schools_df.dtypes
```

```python
# Figure out which programs are 1 day programs versus Semester Long Programs
schools_df['Total Days of Program'] = (schools_df['End Date'] - schools_df['Start Date']).dt.days
schools_df
```

```python
len(schools_df)
```

```python
# See if there are programs that only exist for 1 day - These would be 
# activities like a field trip or surf club and we would not want to include them 
# in our overall Surfing 101 High School Course, we would want them to be separate
surfing_101_to_field_trip = schools_df.loc[(schools_df["Total Days of Program"] == 0) & (schools_df["Activity Name"]=="Surfing 101")]
surfing_101_to_field_trip
```

```python
# Activities with only one day and named SILT

to_field_trip = schools_df.loc[(schools_df["Total Days of Program"] == 0) & (schools_df["Activity Name"]=="Surfers In Leadership Training (SILT)")].index
to_field_trip

# Decided to Drop the one day SILT training since it is a CPR training, not a surfing activity

schools_df = schools_df.drop(to_field_trip)
schools_df
```

```python
also_to_field_trip = schools_df.loc[(schools_df["Total Days of Program"] == 0) & (schools_df["Activity Name"]=="Intro to Surfing Field Trips")]
also_to_field_trip.head()
```

```python
# Want to change the Activity Name based on this condition Total days of program= 0
# Need to create variables to pass through so we can change Activity Names of programs with 0 days of program to Surf Field Trip

#Solution - drop all rows that have total days = 0, create a new dataframe with all activities that have 0 days, then 
#will merge this new dataframe to original 

# Find all activities that were for one day (total days = 0)
one_time_true = schools_df.loc[schools_df["Total Days of Program"] == 0]
one_time_true 



```

```python
# If Activity Name is Surfing 101 change it to Intro to Surfing Field Trip and convert these rows into a new DF
activites_zero = one_time_true.replace(regex={"Surfing 101": "Intro to Surfing Field Trips"})
intro_to_surfing_field_trips_df = pd.DataFrame(activites_zero)
intro_to_surfing_field_trips_df
```

```python
# Drop rows that have one day (total days = 0) from original DF

drop_one_time_true = schools_df.loc[schools_df["Total Days of Program"] == 0].index

schools_df = schools_df.drop(drop_one_time_true)

schools_df
```

```python
intro_to_surfing_field_trips_df
```

```python
schools_df
```

```python
clean_schools_df = pd.concat([schools_df, intro_to_surfing_field_trips_df])
clean_schools_df
```

```python
#Confirm that union worked and kept all columns 
# 71 is correct because we dropped one SILT course
len(clean_schools_df)
```

```python
clean_schools_df["Activity Name"].value_counts()
```

```python
# Check to see if all Surfing PE courses are in High schools
pe = clean_schools_df["Activity Name"]=="Surfing 101 PE Course"
clean_schools_df[pe]
```

```python
# Change Surfing 101 PE Course to Surfing 101 and then change all Surfing 101 to Surfing 101/PE 
# Create list for what we want to change

clean_schools_df["Activity Name"] = clean_schools_df["Activity Name"].replace({"Surfing 101": "Surfing 101 HS/PE Course", "Surfing 101 PE Course": "Surfing 101 HS/PE Course"})
clean_schools_df["Activity Name"].value_counts()
```

## Move DF to PgAdmin


```python
from sqlalchemy.types import Integer, Text, String, DateTime

#Load Data 
db_string = f"postgres://postgres:{db_password}@127.0.0.1:5432/City_Surf"
engine = create_engine(db_string)
clean_schools_df.to_sql(name='schools', con=engine, if_exists = 'replace', dtype={
                  
"Program Name": Text,
"Fiscal Year": Text,
"Activity Name": Text,
"Activity Category": Text,
"Activity Instance ID": Integer,
"Activity Label": Text,
"Service Site Name": Text,
"School": Text,
"Service Site Address": Text,
"Start Date": DateTime,
"End Date": DateTime,
"Days of Week": Text,
"Time": Text,
"Total Days of Program": Integer})
```
